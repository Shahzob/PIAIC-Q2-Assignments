```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv('/content/sample_data/compresive_strength_concrete.csv')
data.head()
```

```python
#Centralize Values
def toCentralizeInput(_data):
  for col in _data.columns:
    if(col != 'Concrete compressive strength(MPa, megapascals) '):
      _data[col] = [x - data[col].mean()/_data[col].std() for x in _data[col]]

toCentralizeInput(data)
data.head()
```

```python
train_X = data.sample(frac=0.8,random_state=0)
val_X = data.drop(train_X.index)
test_X = train_X.sample(frac=0.375,random_state=0)

train_Y = train_X['Concrete compressive strength(MPa, megapascals) ']
train_X.drop('Concrete compressive strength(MPa, megapascals) ',axis=1,inplace=True)

val_Y = val_X['Concrete compressive strength(MPa, megapascals) ']
val_X.drop('Concrete compressive strength(MPa, megapascals) ',axis=1,inplace=True)

test_Y = test_X['Concrete compressive strength(MPa, megapascals) ']
test_X.drop('Concrete compressive strength(MPa, megapascals) ',axis=1,inplace=True)
```

```python
train_X = train_X.values
train_Y = train_Y.values

test_X = test_X.values
test_Y = test_Y.values

val_X = val_X.values
val_Y = val_Y.values
```

```python
from keras import models
from keras import layers

model = models.Sequential()
model.add(layers.Dense(10, activation='relu',
                       input_shape=(train_X.shape[1],)))
model.add(layers.Dense(8, activation='relu'))
model.add(layers.Dense(6, activation='relu'))
model.add(layers.Dense(1))
model.compile(optimizer='rmsprop', loss='mse', metrics=['mae'])
```

```python
model.fit(train_X, train_Y,
              epochs=100, batch_size=10)
```

```python
val_mse, val_mae = model.evaluate(val_X, val_Y)
```

```python
test_mse,test_mae = model.evaluate(test_X,test_Y)
```
