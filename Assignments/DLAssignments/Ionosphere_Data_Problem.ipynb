```python
import numpy as np;
import pandas as pd;
```

```python
data = pd.read_csv('/content/sample_data/ionosphere_data.csv')
```

```python
data.head()
```

```python
data.drop('feature2',axis=1,inplace=True)
data['label'].replace({'b':0,'g':1},inplace=True)
data.info()
```

```python
#data['feature3'] = [x-data['feature3'].mean()/data['feature3'].std() for x in data['feature3']]

#Centralize Values
def toCentralizeInput(_data):
  for col in _data.columns:
    if(col != 'feature1' and col!='label'):
      _data[col] = [x - data[col].mean()/_data[col].std() for x in _data[col]]

toCentralizeInput(data)
data.head()
```

```python
#Train Data
train_X = data.sample(frac=0.6,random_state=0)
test_X = data.drop(train_X.index)

train_Y = train_X['label']
train_X.drop('label',axis=1,inplace=True)

#test data
test_Y = test_X['label']
test_X.drop('label',axis=1,inplace=True)

test_X
```

```python
from keras import models
from keras import layers

model = models.Sequential()
model.add(layers.Dense(16,activation='relu',input_shape=(211,33)))
model.add(layers.Dense(1,activation='sigmoid'))
```

```python
model.compile(optimizer='rmsprop',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
train_X = train_X.values
```

```python
train_X
```

```python
train_Y = train_Y.values
```

```python
model.fit(train_X, train_Y, epochs=100, batch_size=20)
results = model.evaluate(train_X, train_Y)
```
